# 2.categorical column convert to numerical column

```python
# categorical column---->nominal data--->one-hot-encoding
# ordinal data---->label encoder
```

```python
# Handle categorical variables using Label Encoding and One-Hot Encoding
#Label Encoding for 'Area Type', 'Furnishing Status', 'Tenant Preferred'
label_encoder = LabelEncoder()
dataset['Area Type'] = label_encoder.fit_transform(dataset['Area Type'])
dataset['Furnishing Status'] = label_encoder.fit_transform(dataset['Furnishing Status'])
dataset['Tenant Preferred'] = label_encoder.fit_transform(dataset['Tenant Preferred'])
# One-Hot Encoding for 'City' (cities could have many unique values)
dataset = pd.get_dummies(dataset,columns=['City'], drop_first=True,dtype=int)
```

```python
# Drop 'Area Locality' column
dataset.drop('Area Locality', axis=1, inplace=True)
```

```python
dataset
```

```python
# Handle 'Floor' column (splitting 'Floor' into 'Floor Number' and 'Total Floors')
dataset[['Floor Number', 'Total Floors']] = dataset['Floor'].str.split(' out of ', expand=True)
dataset['Floor Number'] = dataset['Floor Number'].astype(int)
dataset['Total Floors'] = dataset['Total Floors'].astype(int)
# Drop the original 'Floor' column
dataset.drop('Floor', axis=1, inplace=True)
```
