```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv("preprocessed_house_rent.csv")
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[[ 'BHK', 'Size', 'Area Type', 'Furnishing Status',
       'Tenant Preferred', 'Bathroom', 'City_Chennai', 'City_Delhi',
       'City_Hyderabad', 'City_Kolkata', 'City_Mumbai', 'Floor Number',
       'Total Floors']]
```

```python
independent
```

```python
dependent=dataset[['Rent']]
```

```python
dependent
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
independent = sc.fit_transform(independent)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
param_grid={'fit_intercept':[True,False],'copy_X':[True,False],'n_jobs':[1,-1],'positive':[False,True]}
grid=GridSearchCV(LinearRegression(),param_grid,refit=True,cv=5,verbose=3,n_jobs=-1)
regressor=grid.fit(independent,dependent)
y_pred=regressor.predict(independent)
y_pred
```

```python
re=grid.cv_results_
grid_predictions = grid.predict(independent)
from sklearn.metrics import r2_score
r_score=r2_score(dependent,grid_predictions)
print("The R_score value for best parameter {}:".format(grid.best_params_),r_score)
```
