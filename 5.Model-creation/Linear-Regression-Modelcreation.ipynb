```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv("preprocessed_house_rent.csv")
```

```python
dataset
```

```python
independent=dataset[[ 'BHK','Size', 'Area Type', 'Furnishing Status',
       'Tenant Preferred', 'Bathroom', 'City_Chennai', 'City_Delhi',
       'City_Hyderabad', 'City_Kolkata', 'City_Mumbai', 'Floor Number',
       'Total Floors']]
```

```python
independent
```

```python
dependent=dataset[['Rent']]
```

```python
dependent
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
independent=sc.fit_transform(independent)
```

```python
# Create and train the model on the entire dataset
model = LinearRegression()
model.fit(independent,dependent)
```

```python
#Predict on the same data (note: this is just for training check)
y_pred = model.predict(independent)
```

```python
# Print R2 score
from sklearn.metrics import r2_score
print("R2 Score (on full data):", r2_score(dependent, y_pred))
```
