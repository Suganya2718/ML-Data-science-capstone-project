```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv("House_rent.csv")
```

```python
df
```

```python
# 2. Check for missing values
df.isnull().sum()
```

```python
# Clean the 'Floor' column (split into 'Current Floor' and 'Total Floors')
def extract_floors(floor_str):
    try:
        if 'Ground' in floor_str:
            return 0, int(floor_str.split('out of')[-1].strip())
        else:
            parts = floor_str.split('out of')
            return int(parts[0].strip()), int(parts[1].strip())
    except:
        return np.nan, np.nan

df[['Current Floor', 'Total Floors']] = df['Floor'].apply(lambda x: pd.Series(extract_floors(str(x))))
```

```python
# Drop original floor column
df.drop('Floor', axis=1, inplace=True)
```

```python
# Drop unnecessary columns (optional)
df.drop('Area Locality', axis=1, inplace=True)
```

```python
# One-hot encode categorical variables
categorical_cols = ['Area Type', 'City', 'Furnishing Status', 'Tenant Preferred']
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=False,dtype=int)
```

```python
# Save the cleaned dataset to a new CSV
df_encoded.to_csv('cleaned_house_rent_dataset.csv', index=False)
```
